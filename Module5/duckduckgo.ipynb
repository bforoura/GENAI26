# **DuckDuckGo**


* The **DuckDuckGoSearchRun** tool is a **wrapper** for integration into **LLM agents**.

* The primary attributes and parameters used in the example:

   * **name** is the internal identifier for the tool; e.g., **duckduckgo_search**

   * **description** is an explanation of what the tool does. This is used by an LLM to decide when to trigger a search.

   * **args_schema** defines the input structure; it expects a single string for the search query

   * **run(query)**: The primary method to execute the search.

```python
from langchain_community.tools import DuckDuckGoSearchRun

# Initialize the tool
search = DuckDuckGoSearchRun()

# Print the tool information from your example
print(f"Tool's name = {search.name}")
print(f"Tool's description = {search.description}")
print(f"Tool's arg schema = {search.args_schema}")

print("-" * 30)

# Run a simple search
result = search.run("Current weather in Philadelphia")
print(result)

```
