```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper

# Initialize the API Wrapper
# top_k_results=1 means it will only grab the most relevant article
# doc_content_chars_max=1000 limits the length of the text
api_wrapper = WikipediaAPIWrapper(top_k_results=1, doc_content_chars_max=1000)



# Initialize the tool
wiki = WikipediaQueryRun(api_wrapper=api_wrapper)



# Print the tool parameters
print(f"Tool's name        = {wiki.name}")
print(f"Tool's description = {wiki.description}")
print(f"Tool's arg schema  = {wiki.args_schema}")

print("-" * 30)



# Run a simple search
# The result will be a summary of the Wikipedia page for "Philadelphia"
result = wiki.run("Philadelphia")
print(result)

```
